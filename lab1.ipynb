```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x,y ,label='sin(x)')
plt.title('Test')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,15))
plt.subplot(2,2,1)
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.title("Sin vs Cos")
plt.legend()

plt.subplot(2,2,2)
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, c='blue', alpha = 0.5)
plt.title('Scatter Plot')


plt.subplot(2,2,3)
categories = ['A', 'B', 'C', 'D']
values = np.random.rand(4) * 50
plt.bar(categories, values)
plt.title('Bar Plot')

plt.subplot(2,2,4)
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.title('Histogram')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as py 

x = np.random.randn(50)
y = x * np.random.randn(50) * 10
sizes = np.random.rand(50) * 200 + 10

plt.figure(figsize=(10,15))
scatter = plt.scatter(x,y, s=sizes, c=sizes, cmap='viridis', alpha=0.6)
plt.colorbar(scatter, label="Variable Size")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.show()
```
