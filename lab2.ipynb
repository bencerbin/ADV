```python
import matplotlib.pyplot as plt
import numpy as np

def create_static_plots():
    x = np.linspace(0, 10, 100)
    y1 = np.cos(x)
    y2 = np.sin(x)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,6))
    ax1.plot(x, y1, label='cos(x)')
    ax1.plot(x, y2, label='sin(x)')
    ax1.set_title("Plot 1")
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    ax2.scatter(x[::5], y1[::5], label='cos(x)', alpha=0.6)
    ax2.scatter(x[::5], y2[::5], label='sin(x)', alpha=0.6)
    ax2.set_title("Plot 2")
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()
    
create_static_plots()
```

```python
def create_statistical_plots():
    # Create sample data
    categories = ['A', 'B', 'C', 'D']
    values = np.random.randint(10, 50, size=len(categories))
    
    # Create random data for box plots
    data = [np.random.normal(30, 5, 100), 
            np.random.normal(20, 2, 100),
            np.random.normal(40, 8, 100)]
    
    # Create figure
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Bar plot
    ax1.bar(categories, values, color='skyblue')
    ax1.set_title('Bar Plot')
    ax1.set_xlabel('Categories')
    ax1.set_ylabel('Values')
    
    # Box plot
    ax2.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'])
    ax2.set_title('Box Plot')
    ax2.set_ylabel('Values')
    
    plt.tight_layout()
    plt.show()

create_statistical_plots()
```

```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd

def create_interactive_plots():
    x = np.linspace(0, 10, 100)
    y1 = np.sin(x)
    y2 = np.cos(x)
    
    df = pd.DataFrame({
        'x': np.concatenate([x,x]),
        'y': np.concatenate([y1,y2]),
        'type': ['sin']*100 + ['cos']*100
    })

    fig1 = px.line(df, x='x', y='y', color='type', title='Interactive Line Plot')
    fig1.show()
    
    fig2 = px.scatter(df, x='x', y='y', color='type', title='Interactive Line Plot')
    fig2.show()
    
create_interactive_plots()
    
```

```python
def create_statistical_plots():
    # Create sample data
    categories = ['A', 'B', 'C', 'D']
    values = np.random.randint(10, 50, size=len(categories))
    
    data = [
        np.random.normal(30, 5, 100),  # Mean 30, std dev 5
        np.random.normal(20, 2, 100),  # Mean 20, std dev 2
        np.random.normal(40, 8, 100)   # Mean 40, std dev 8
    ]
    
    fig1 = go.Figure(data = [go.Bar (
        x = categories,
        y = values,
        marker_color = 'skyblue',
        text = values,
        textposition = 'auto'
        
    )])
    
    fig1.show()
    
    df_box = pd.DataFrame({
        'Values' : np.concatenate(data),
        'Group': ['Group 1'] * 100 + ['Group 2'] * 100 + ['Group 3'] *100
        
    })
    
    fig2 = px.box(df_box, x='Group', y='Values', title='Box Plot')
    fig2.show()
    

create_statistical_plots()
```
